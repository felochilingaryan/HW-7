```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
```

```python
browser = webdriver.Chrome()
url = "http://rate.am/en/armenian-dram-exchange-rates/banks/non-cash"
browser.get(url)
```

```python
data=browser.find_element_by_xpath('//*[@id="aspnetForm"]/div[3]/div[2]/div[1]/div[4]/div[3]/table/tbody/tr/td[1]/a')
data.click()
data_1=browser.find_element_by_xpath('//*[@id="calBorder"]/span/table/tbody/tr[3]/td/table/tbody/tr[5]/td[4]')
data_1.click()
data_2=browser.find_element_by_xpath('//*[@id="ctl00_Content_RB_dtpick1_btnOk"]')
data_2.click()
```

```python
import pandas as pd
```

```python
rates_table=pd.read_html(browser.find_element_by_class_name("rb").get_attribute('outerHTML'))
rates_table = rates_table[0]
rates_table = rates_table[2:19] 
del rates_table[0]
del rates_table[2]
del rates_table[3] 
```

```python
rates_table.columns = ["Բանկ", "Ամսաթիվ", "USD_Առք", "USD_Վաճ", "EUR_Առք", "EUR_Վաճ", "RUB_Առք", "RUB_Վաճ", "GBF_Առք", "GBF_Վաճ"]
```
